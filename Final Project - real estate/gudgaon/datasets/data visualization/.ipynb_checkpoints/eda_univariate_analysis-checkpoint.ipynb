#### Descriptive Statistics 
-  Count: There are 3704 non-missing price entrie.
- Mean Price: The average price is approximate 2.5353 crore
- Median Price: The median (or 50th percentile) price 1.521.52 cror
- Standard Deviation: The prices have a standard deviation of 2.98, indicating variability in the price
- Range: Prices range from a minimum of 0.07 crores to a maximum of 31.5 crores.
- IQR: The interquartile range (difference between 75th and 25th percentile) is from 0.95 crores to 2.75 c
  
#### Visualization
- Distribution: The histogram indicates that most properties are priced in the lower range (below 5 crores), with a few properties going beyond 
10 crores
- Box Plot: The box plot showcases the spread of the data and potential outliers. Properties priced above approximately 10 crores might be -considered outliers as they lie beyond the upper whisker of the box plot.
- Missing Values: There are 16 missing values in the price column. ice column.

```python
# lets check skewness and kurtosis
skewness = df.price_cr.skew()
kurtosis = df.price_cr.kurt()
print(skewness,kurtosis)
```

- Skewness: The price distribution has a skewness of approximately 3.28, indicating a positive skew. This means that the distribution tail is skewed to the right, which aligns with our observation from the histogram where most properties have prices on the lower end with a few high-priced properties.
- 
Kurtosis: The kurtosis value is approximately 14.93. A kurtosis value greater than 3 indicates a distribution with heavier tails and more outliers compared to a normal distribution.

```python
# quantile analysis
df.price_cr.quantile([0.01,0.05,0.25,0.50,0.75,0.95,0.99])
```

Quantile Analysis:
- 1% Quantile: Only 1% of properties are priced below 0.25 crores.
- 5% Quantile: 5% of properties are priced below 0.37 crores.
- 95% Quantile: 95% of properties are priced below 8.5 crores.
- 99% Quantile: 99% of properties are priced below 15.097 crores, indicating that very few properties are priced above this value.
ralue.

```python
# identify potential outliers using IQR method 
Q1 = df.price_cr.describe()['25%']
Q3 = df.price_cr.describe()['75%']
IQR = Q3 - Q1

IQR
```

```python
lower_bound =  Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound, upper_bound)
```

```python
outliers = df[(df.price_cr < lower_bound) | (df.price_cr > upper_bound)]
outliers
# there are too may outliers and we can't remove them 
```

```python
outliers.price_cr.describe()
```

#### outlier Analysis(using IQR method):

- there are 431 properties considered as an outliers.
- These outliers have avg price of approx 9.20 CR.
- The range of outliers range from 5.46 to 31.5 cr

```python
# price binning 
bins = [0,1,2,3,5,10,20,50]

bin_lables = ["0-1", "1-2", "2-3", "3-5", "5-10", "10-20", "20-50"]

pd.cut(df.price_cr,bins = bins, labels = bin_lables,right=False).value_counts().sort_index().plot(kind='bar')
pd.cut(df.price_cr,bins = bins, labels = bin_lables,right=False).value_counts().sort_index()
```

```python
# ecdf plot 
ecdf = df.price_cr.value_counts(normalize=True).sort_index().cumsum()
plt.plot(ecdf.index,ecdf,marker = '.',linestyle='none')
plt.grid()
```

```python
df.price_cr.value_counts(normalize=True).sort_index()
```

```python
# Price data is right skewed and have large amount of outliers 
# We can apply LOG to reduce outliers for RIGHT SKEWED and to make graph more normalize
plt.figure(figsize=(15, 6))

# Distribution plot without log transformation
plt.subplot(1, 2, 1)
sns.histplot(x=df['price_cr'], kde=True, bins=50, color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.histplot(x=np.log1p(df['price_cr']), kde=True, bins=50, color='lightgreen') 
# as value of price is smaller (less than 0) , hence using LOG1P which gives natural log of (1 + X)
plt.title('Distribution of Prices (Log Transformed)')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python

(np.log1p(df.price_cr)).skew()
```

```python
(np.log1p(df.price_cr)).kurt()
```

```python
plt.figure(figsize=(15, 6))

# Distribution plot without log transformation
plt.subplot(1, 2, 1)
sns.boxplot(x=df['price_cr'], color='skyblue')
plt.title('Distribution of Prices (Original)')
plt.xlabel('Price (in Crores)')
plt.ylabel('Frequency')

# Distribution plot with log transformation
plt.subplot(1, 2, 2)
sns.boxplot(x=np.log1p(df['price_cr']), color='lightgreen')
plt.title('Distribution of Prices (Log Transformed)')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
