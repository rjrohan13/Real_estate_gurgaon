```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("oulier_detection.csv")
```

```python
df['price_per_sqft'].describe()
```

```python
df[df['area']/df['bedRoom']<253]
```

```python
sns.lmplot(data=df,x='area',y='bedRoom')
```

```python
df['area_room_ratio']=df['area']/df['bedRoom']
```

```python
(df[df['area_room_ratio']<253])['bedRoom'].describe()
```

```python
df=df[df['area_room_ratio']>100]
```

```python
outliers_df=df[(df['area_room_ratio']<253) & (df['bedRoom']>3)]
```

```python
outliers_df.sample(5)
```

```python
outliers_df.columns
```

```python
outliers_df['bedRoom']=round(outliers_df['bedRoom']/outliers_df['floorNum'])
```

```python
df.update(outliers_df)
```

```python
df['area_room_ratio']=df['area']/df['bedRoom']
```

```python
df[(df['area_room_ratio']<253) & (df['bedRoom']>4)]
```

```python
df=df[~((df['area_room_ratio']<253) & (df['bedRoom']>4))]
```

```python
sns.lmplot(data=df,x='area',y='bedRoom')
```

```python
df.to_csv('final_outlier_detection.csv')
```
